De la variable **temperatura promedio por año**, ordené los valores de manera ascendente y obtuvé sus cuartiles para formar 4 grupos tomando en cuenta los países a los que pertenecen esas valores. Es decir:

- El grupo "a" lo conforman aquellos países que tienen una temperatura menor a 11 grados. 

- El grupo "b" aquellos países con una temperatura entre los 11 y 22 grados.

- El grupo "c" lo constituyen aquelllos países que tienen una temperatura entre los 22 y 25 grados.

- El grupo "d" aquellos países con temperaturas superiores a los 25 grados.

```python
# Librerías
import pandas as pd
import pdpipe as pdp 
import seaborn as sns
import numpy as np
import plotly.graph_objs as go
import plotly.express as px
from plotly.subplots import make_subplots
from pandas_profiling import ProfileReport
import matplotlib.pyplot as plt
import io
from plotly.offline import plot
```

```python
# Datos que corresponden a los casos confirmados por covid-19
df = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
df = df.rename(columns={'Country/Region': 'Country'})
df.rename(columns={'Country/Region':'Country'}, inplace=True)
df = df.drop(["Province/State", "Lat", "Long"], axis="columns")
df = df.set_index('Country')
df = df.stack()
df.index = df.index.rename('Date', level=1)
df.name = 'Cases'
df = df.reset_index()
```

```python
# Grupo de países con temperaturas inferiores a 11 grados en promedio por año.
group_a = df.loc[df['Country'].isin(['Canada', 'Russia', 'Mongolia', 'Kyrgyzstan', 'Sweden',
                                     'US', 'China', 'Chile', 'Turkey', 'Kazakhstan', 
                                     'Japan', 'Ukraine', 'Poland', 'Romania', 'Germany', 'France', 
                                     'United_Kingdom', 'New_Zealand', 'Georgia', 'Armenia', 'Bulgaria', 
                                     'Belarus', 'Latvia', 'Serbia', 'Croatia', 'Slovakia', 'Hungary', 
                                     'Austria', 'Czechia', 'Switzerland', 'Montenegro', 'Kosovo', 
                                     'North_Macedonia', 'Bhutan', 'Moldova', 'Denmark', 'Lithuania', 
                                     'Bosnia_and_Herzegovina', 'Slovenia', 'Belgium', 'Netherlands', 'Ireland', 
                                     'Falkland_Islands_(Malvinas)', 'Luxembourg', 'Andorra', 'Isle_of_Man', 
                                     'Liechtenstein'])]

fig = px.line(group_a, x = "Date", y = "Cases", text = "Cases", color = "Country")
fig.show()
```

En la gráfica anterior podemos observar que sólo uno de estos países sobresale en la gráfica: Estados Unidos con 905 358 casos confirmados hasta el día 24 de abril del año en curso. El resto de los países presentan casos menores a 160 000.

```python
group_b = df.loc[df['Country'].isin(['Australia', 'Mexico', 'Argentina', 'Libya', 'Angola', 'South_Africa', 
                                     'Pakistan', 'Afghanistan', 'Iran', 'Uzbekistan', 'Peru', 'Portugal', 
                                     'Spain', 'Morocco', 'Italy', 'Tunisia', 'Egypt', 'Jordan', 'Syria', 'Iraq', 
                                     'Malawi', 'Zambia', 'Namibia', 'Botswana', 'Zimbabwe', 'Nepal', 'Bolivia', 
                                     'Greece', 'Israel', 'Taiwan', 'South_Korea', 'Azerbaijan', 'Ecuador', 
                                     'Uruguay', 'Cape_Verde', 'Albania', 'Cyprus', 'Lebanon', 'New_Caledonia', 
                                     'San_Marino'])]

fig = px.line(group_b, x = "Date", y = "Cases", text = "Cases", color = "Country")
fig.show()
```

En el grupo B, países con temperatura entre los 11 y 22 grados, la mayoría se encuentran abajo de 23 000 casos, excepto los países de: Irán con 88 194, Italia con 192 994 y España con 219 764, éste último presenta una pendiente pronunciada, lo cual nos indica que crece de manera acelerada. 

En el grupo A, los casos que se presentan son menores a 160 000 y los del grupo B se acerca un poco a los 23 000.

```python
group_c = df.loc[df['Country'].isin(['Indonesia', 'India', 'Brazil', 'Algeria', 'Ethiopia', 'Kenya', 
                                     'Democratic_Republic_of_the_Congo', 'Mozambique', 'Madagascar', 'Myanmar', 
                                     'Vietnam', 'Saudi_Arabia', 'French_Polynesia', 'Colombia', 
                                     'Central_African_Republic', 'Cameroon', 'Uganda', 
                                     'United_Republic_of_Tanzania', 'Congo', 'Laos', 'Yemen', 
                                     'Papua_New_Guinea', 'Paraguay', 'Gabon', 'Fiji', 'Guatemala', 'Honduras', 
                                     'Timor_Leste', 'Kuwait', 'Bahamas', 'Dominican_Republic', 'Haiti', 
                                     'Jamaica', 'Belize', 'El_Salvador', 'Costa_Rica', 'Sao_Tome_and_Principe', 
                                     'Equatorial_Guinea', 'Puerto_Rico', 'Mauritius', 'Dominica'])]

fig = px.line(group_c, x="Date", y="Cases", text="Cases", color="Country")
fig.show()
```

El grupo C, países con temperatura entre los 22 y 25 grados, presentan al rededor de 8 000 casos. Sólo India tiene 24 530 y Brasil, 54 043. 

La mayoría de los países en este grupo presentan un número de casos menores a los grupos anteriores. Hasta este momento podemos observar que conforme aumenta la temperatura, el número de casos presentados disminuye.

```python
group_d = df.loc[df['Country'].isin(['Sudan', 'Chad', 'Malaysia', 'Somalia', 'Ghana', 'Nigeria', 'Eritrea', 
                                     'South_Sudan', 'Thailand', 'Philippines', 'Oman', 'Venezuela', 'Guyana', 
                                     'Cuba', 'Guinea', 'Liberia', 'Togo', 'Cambodia', 'Bangladesh', 'Sri_Lanka', 
                                     'Suriname', 'Nicaragua', 'Panama', 'Guinea_Bissau', 'Sierra_Leone', 
                                     'Brunei_Darussalam', 'Qatar', 'Trinidad_and_Tobago', 'Singapore'])]

fig = px.line(group_d, x="Date", y="Cases", text="Cases", color="Country")
fig.show()
```

En el grupo D, países con temperaturas mayores a 25 grados, una gran parte se encuentran con números inferiores a 1 790. Lo que se concluye hasta este momento es que, conforme la temperatura aumenta el número de casos disminuye. E incluso los países con un caso extremo, disminuyen de la misma manera.

```python
# Datos que corresponden a los casos confirmados por covid-19
df = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
df = df.rename(columns={'Country/Region': 'Country'})
df.rename(columns={'Country/Region':'Country'}, inplace=True)
df = df.drop(["Province/State", "Lat", "Long"], axis="columns")
df = df.set_index('Country')
df = df.stack()
df.index = df.index.rename('Date', level=1)
df.name = 'Deaths'
df = df.reset_index()
```

```python
group_ad = df.loc[df['Country'].isin(['Canada', 'Russia', 'Mongolia', 'Kyrgyzstan', 'Sweden',
                                     'US', 'China', 'Chile', 'Turkey', 'Kazakhstan', 
                                     'Japan', 'Ukraine', 'Poland', 'Romania', 'Germany', 'France', 
                                     'United_Kingdom', 'New_Zealand', 'Georgia', 'Armenia', 'Bulgaria', 
                                     'Belarus', 'Latvia', 'Serbia', 'Croatia', 'Slovakia', 'Hungary', 
                                     'Austria', 'Czechia', 'Switzerland', 'Montenegro', 'Kosovo', 
                                     'North_Macedonia', 'Bhutan', 'Moldova', 'Denmark', 'Lithuania', 
                                     'Bosnia_and_Herzegovina', 'Slovenia', 'Belgium', 'Netherlands', 'Ireland', 
                                     'Falkland_Islands_(Malvinas)', 'Luxembourg', 'Andorra', 'Isle_of_Man', 
                                     'Liechtenstein'])]

group_ad_max = group_ad.groupby(['Country']).agg(Deaths_max=('Deaths', 'max')).reset_index()
```

```python
group_bd = df.loc[df['Country'].isin(['Australia', 'Mexico', 'Argentina', 'Libya', 'Angola', 'South_Africa', 
                                     'Pakistan', 'Afghanistan', 'Iran', 'Uzbekistan', 'Peru', 'Portugal', 
                                     'Spain', 'Morocco', 'Italy', 'Tunisia', 'Egypt', 'Jordan', 'Syria', 'Iraq', 
                                     'Malawi', 'Zambia', 'Namibia', 'Botswana', 'Zimbabwe', 'Nepal', 'Bolivia', 
                                     'Greece', 'Israel', 'Taiwan', 'South_Korea', 'Azerbaijan', 'Ecuador', 
                                     'Uruguay', 'Cape_Verde', 'Albania', 'Cyprus', 'Lebanon', 'New_Caledonia', 
                                     'San_Marino'])]

group_bd_max = group_bd.groupby(['Country']).agg(Deaths_max=('Deaths', 'max')).reset_index()
```

```python
group_cd = df.loc[df['Country'].isin(['Indonesia', 'India', 'Brazil', 'Algeria', 'Ethiopia', 'Kenya', 
                                     'Democratic_Republic_of_the_Congo', 'Mozambique', 'Madagascar', 'Myanmar', 
                                     'Vietnam', 'Saudi_Arabia', 'French_Polynesia', 'Colombia', 
                                     'Central_African_Republic', 'Cameroon', 'Uganda', 
                                     'United_Republic_of_Tanzania', 'Congo', 'Laos', 'Yemen', 
                                     'Papua_New_Guinea', 'Paraguay', 'Gabon', 'Fiji', 'Guatemala', 'Honduras', 
                                     'Timor_Leste', 'Kuwait', 'Bahamas', 'Dominican_Republic', 'Haiti', 
                                     'Jamaica', 'Belize', 'El_Salvador', 'Costa_Rica', 'Sao_Tome_and_Principe', 
                                     'Equatorial_Guinea', 'Puerto_Rico', 'Mauritius', 'Dominica'])]

group_cd_max = group_cd.groupby(['Country']).agg(Deaths_max=('Deaths', 'max')).reset_index()
```

```python
group_dd = df.loc[df['Country'].isin(['Sudan', 'Chad', 'Malaysia', 'Somalia', 'Ghana', 'Nigeria', 'Eritrea', 
                                     'South_Sudan', 'Thailand', 'Philippines', 'Oman', 'Venezuela', 'Guyana', 
                                     'Cuba', 'Guinea', 'Liberia', 'Togo', 'Cambodia', 'Bangladesh', 'Sri_Lanka', 
                                     'Suriname', 'Nicaragua', 'Panama', 'Guinea_Bissau', 'Sierra_Leone', 
                                     'Brunei_Darussalam', 'Qatar', 'Trinidad_and_Tobago', 'Singapore'])]

group_dd_max = group_dd.groupby(['Country']).agg(Deaths_max=('Deaths', 'max')).reset_index()
```

```python
groups_A = pd.concat([group_a_max, group_ad_max], axis = 1)
groups_A.columns = ['Country', 'Cases_max', 'Country_2', 'Deaths_max']
groups_A = groups_A.drop(columns=['Country_2'])

groups_B = pd.concat([group_b_max, group_bd_max], axis = 1)
groups_B.columns = ['Country', 'Cases_max', 'Country_2', 'Deaths_max']
groups_B = groups_B.drop(columns=['Country_2'])

groups_C = pd.concat([group_c_max, group_cd_max], axis = 1)
groups_C.columns = ['Country', 'Cases_max', 'Country_2', 'Deaths_max']
groups_C = groups_C.drop(columns=['Country_2'])

groups_D = pd.concat([group_d_max, group_dd_max], axis = 1)
groups_D.columns = ['Country', 'Cases_max', 'Country_2', 'Deaths_max']
groups_D = groups_D.drop(columns=['Country_2'])
```

```python
groups_A = groups_A.assign(num = groups_A.Cases_max.astype(int))
groups_B = groups_B.assign(num = groups_B.Cases_max.astype(int))
groups_C = groups_C.assign(num = groups_C.Cases_max.astype(int))
groups_D = groups_D.assign(num = groups_D.Cases_max.astype(int))
cov_a = ['Country', 'Cases_max', 'Deaths_max']
```

En la gráfica paralela que se presenta a continuación que representa al grupo A, tenemos 3 barras principales y la barra de color presentada de manera horizontal. En la primer barra se encuentran los nombres de los países que pertenecen al primer grupo, en la segunda el número de casos de cada uno de esos países, y en la tercera el número de defunciones. Mongolia, presenta 37 casos y ninguna muerte. Liechtenstein presenta más del doble que Mongolia (81 casos confirmados) y sólo una muerte. Estados Unidos presenta 51 949 muertes hasta este momento.

En la segunda gráfica se presenta una Scatter entre el número de casos confirmados y el número de muertes. Sobresale Estados Unidos con un número grande casos y así mismo un número representativo de muertes.

```python
fig1 = px.parallel_categories(groups_A, dimensions = cov_a, color = 'num')
fig2 = px.scatter(groups_A, x='Deaths_max', y='Cases_max', color='Country')
fig1.show()
fig2.show()
```

```python
fig1 = px.parallel_categories(groups_B, dimensions = cov_a, color = 'num')
fig2 = px.scatter(groups_B, x='Deaths_max', y='Cases_max', color='Country')
fig1.show()
fig2.show()
```

```python
fig1 = px.parallel_categories(groups_C, dimensions = cov_a, color = 'num')
fig2 = px.scatter(groups_C, x='Deaths_max', y='Cases_max', color='Country')
fig1.show()
fig2.show()
```
